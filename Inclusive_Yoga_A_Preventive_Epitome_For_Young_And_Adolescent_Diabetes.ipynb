```python
#Headers
import pandas as pd
import numpy as np
import matplotlib as mp
import sklearn as sk
import seaborn as sb
import os
from PIL import Image
import time
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt

def loadfile():
    #Accepting input file 
    filename = "C:/Users/yukta/Desktop/trainfile.csv"
    #Loading the dataset to dataframe 
    df = pd.read_csv(filename)
    #View file
    #print(df)
    print("Function loadfile executed")
    return df

def setindex(df):
    df1 = df['c1']
    df.set_index('c1', inplace=True)
#     print(df)
#     print(df1)
    print("Function setindex executed")
    return df,df1

def changedatatype(df):
    df = df.astype(float)
    #print(df.dtypes)
    print("Function changedatatype executed")
    return df

def viewinfo(df):
    #View info
    df.info()
    #Data Science Operations
    print('No.of columns',df.shape[1])
    print('No.of rows',df.shape[0])
    print("Function viewinfo executed")
    return df

def checkempty(df):
    #Check if there is any empty cells
    df.empty
    #Fill the empty columns using pandas
    df = df.fillna(99)
    #print(df)
    print("Function checkempty executed")
    return df

def traintestsplit(df):
    rowmask = df.isnull().any(axis=1)
    columnmask = df.isnull().any(axis=0)
    #Test Train Splitter 
    X = df[df.columns[0:-1]]
    y = df[df.columns[-1]]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
#     print('X train data is ',X_train)
#     print('Y test data is ',y_test)
    print("Function traintestsplit executed")
    return X_train, X_test, y_train, y_test

def decisiontree(X_train, X_test, y_train, y_test):
    #Train - f3
    Tree = DecisionTreeClassifier(criterion = 'gini')
    Tree.fit(X_train, y_train)
    #Test - f3
    predictions = Tree.predict(X_test)
    print('Accuracy on Training set: ',Tree.score(X_train,y_train))
    print('Accuracy on Test set: ',Tree.score(X_test,y_test))
    print("Function decisiontree executed")
    return predictions, y_test, Tree

def predictions(predictions, y_test, df):
    #Confusion matrix
    print('Confusion matrix: ',confusion_matrix(y_test,predictions),end='\n\n')
    #Accuracy score
    print('Accuracy score: ',accuracy_score(y_test,predictions),end='\n\n')
    print('Outcome: ',df['c62'].value_counts())
    print("Function predictions executed")
    return 0

def testpredictions(Tree):
    testfilename = "C:/Users/yukta/Desktop/testfile.csv"
    df2 = pd.read_csv(testfilename)
    #print(df2.info())
    df3 = df2['c1']
    df2.set_index('c1', inplace=True)
    df2 = df2.astype(float)
    #print(df2.dtypes)

#     print('No.of columns',df2.shape[1])
#     print('No.of rows',df2.shape[0])

    rowmask = df2.isnull().any(axis=1)
    columnmask = df2.isnull().any(axis=0)
    
    testprediction = Tree.predict(df2)
#     print('df2', df2)
    print("Function testpredictions executed")
    return df3, testprediction

def writecsvfile(df3, testprediction):
    
    file = open('C:/Users/yukta/Desktop/Userprediction80.csv','a')
    file.write('UID,FutureDiabetic\n')
    
    for i in range(131):
        file.write(str(df3[i]))
        file.write(',')
        file.write(str(testprediction[i]))
        file.write('\n')
        print(df3[i], testprediction[i], '\n')
    
    file.close()
    print("CSV GENERATED!")
    print("Function writecsvfile executed")
    return 0

 
def displayplots(df,df1):
    #Visualization - plot 
    a = df1
    b = df['c62']
    plt.plot(a,b)
    plt.xlabel('x - axis')
    plt.ylabel('y - axis')
    plt.title('Basic Visualization')
    plt.show()
    #Scatter Plot
    a1 = df1
    b1 = df['c62']
    plt.scatter(a1, b1, label= "stars", color= "green", marker= "*", s=30)
    plt.xlabel('x - axis')
    plt.ylabel('y - axis')
    plt.title('Scatter Plot - Pro-diabetic and non-diabetic')
    plt.legend()
    plt.show()
    print("Function displayplots executed")
    return 0
    
def identifyuser():
    uniqueuserid = int(input('Unique user id'))
    #Open userprediction file which contains only 2 cols UID and future diabetic
    up_file = open('C:/Users/yukta/Desktop/Userprediction80.csv','r')
    #Load userprediction file
    user_var = pd.read_csv(up_file)
#     print('PRINTING USER_VAR', user_var)
#     print('uniqueuserid', uniqueuserid)
#     print('COLUMNS', user_var.columns)
#     print('ILOC FUNC', user_var.iloc[uniqueuserid][1])
    if user_var.iloc[uniqueuserid-1][1] == 1.0 or user_var.iloc[uniqueuserid-1][1] == 99.0:
        subclassfile = "C:/Users/yukta/Desktop/subclassprediction.csv"
        scf_var = pd.read_csv(subclassfile)
        scf_var.empty
        scf_var = scf_var.fillna(99)
        res = scf_var.iloc[uniqueuserid-1][84]
        print(res)
    else:
        res = 0
        print("You are not pro-diabetic, but please do be cautious!")
    return res
            
    
def displayimages(res):
    print("Function displayimages executed")
    file = open('C:/Users/yukta/Desktop/Userprediction80.csv','r')
    folderpath = 'C:/Users/yukta/OneDrive/Documents/Yukta-yoga-project/yogadatasetpictures/'
    if res == 0 or res == '0':
        return 0
    else:
        diabetesprediction = res+'/'
        for i in range(1,3):
            filelocation = folderpath + diabetesprediction + 'file_' + str(i) + '.jpg'
            img = Image.open(filelocation)
            img.show()
            time.sleep(5)
        return 0

def main():
    a1 = loadfile()
    b1,b2 = setindex(a1)
    d1 = changedatatype(b1)
    e1 = viewinfo(d1)
    f1 = checkempty(e1)
    a,b,c,d = traintestsplit(f1)
    p,y,rt = decisiontree(a,b,c,d)
    predictions(p,y,f1)
    d,tp = testpredictions(rt)
    writecsvfile(d,tp)
    #displayplots(f1,b2)
    result = identifyuser()
    displayimages(result)
    
if __name__ == "__main__":
    main()
```
